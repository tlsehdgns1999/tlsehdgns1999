##2. 분류와 회귀에 사용하는 머신러닝 알고리즘 K-최근접 이웃

- k-NN은 사실 다른 기법들 처럼 학습을 한다고 볼 수 없음
- 지역적인 변수를 추출해 게산하고 값을 저장하는 과정을 반복하기 때문임
- 계산량이 상당히 높음
- 통상 게으른 알고리즘, 인스턴스 기반 학습이라고 불리움

###2.1. K-최근접 이웃(분류)의 개념

- k-NN 분류에서 출력은 소속된 항목이다. 객체는 k개의 최근접 이웃 사이에서 가장 공통적인 항목에 할당되는 객체로 과반수 의결에 의해 분류된다. 만약 k = 1 이라면 객체는 단순히 하나의 최근접 이웃의 항목에 할당된다.
- k는 통상 [(2n-1) 단 n>=1] 을 따라간다.

image.png

(위 그림에서 k=3 일땐 빨간색, k=5일 땐 파란색)
-> 최적의 k값 설정이 굉장히 중요하다

- 훈련 데이터는 각각 항목 분류명을 갖는 다차원 특징 공간에서의 벡터임
- 알고리즘의 훈련 단계에선 훈련 표본의 특징 벡터와 항목 분류명을 저장함

- 연속 변수에서 가장 흔하게 사용되는 거리 척도는 "유클리드 거리"
- 문자 분류와 같은 이산 변수의 경우 중첩 거리(또는 해밍 거리)와 같은 다른 척도가 사용될 수 있음
(예를 들어 유전자 관련 데이터의 경우 피어슨과 스피어만 같은 상관 계수를 사용함)
- 종종 거리 척도가 큰 여백 최근접 이웃이나 이웃 성분 분석(비지도 학습의 일종)과 같은 특별한 알고리즘으로 학습된다면 k-NN 분류의 정확성을 상당히 향상시킬 수 있음

- 과반수 의결에 따른 분류의 단점은 항목 분포가 편향될 때 나타남 (빈번한 항목의 데이터가 새로운 데이터의 예측을 지배하는 경향이 있다.)
- 더 빈번한 항목의 데이터가 다수를 이루어서 k개의 최근접 이웃의 대부분이 되는 경향이 있기 때문임
- 이 문제를 해결하는 방법중 하나는 검증점과 k개의 최근접 이웃 각각의 거리를 고려하여 분류에 가중치를 주는 것임
- 각각 k개의 최근접 점의 항목에 그 점에서 검증점까지의 거리에 반비례하는 가중치를 곱함(1/n)
- 편향을 해결하는 또 다른 방법은 데이터 표현을 추상화하는 방법도 존재함

###2.3. 차원의 저주에 취약한 k-최근접 이웃

####2.3.1. 차원의 저주란?

- 차원이 증가할 수록 학습 데이터의 수가 차원(컬럼)의 개수보다 작아져 모델이 제대로 학습을 할 수 없게 만들어지는 것
- 대체로 over fitting 이 발생함

####2.3.2. k-최근접 이웃이 차원의 저주에 빠지는 이유

- 차원이 증가함에 따라 데이터 사이의 빈 공간이 많이 생기게 됨
- 빈 공간이 생긴다는 것은 데이터의 값이 0으로 채워진다는 것을 의미!

##2.4 차원의 저주를 해결하는 C-knn

###2.4.1. C-knn

압축된 근접 이웃(C-knn)은 k-NN 분류를 위해 데이터 집합을 축소하려는 목적으로 설계된 알고리즘이다.이 알고리즘은 훈련 데이터로부터 프로토타입의 집합 U를 선택하고, 이를 통해 1NN이 모든 데이터 집합을 분류하는 것만큼 정확하게 U와 1NN이 데이터들을 분류할 수 있다.

image.png

- X의 모든 요소를 탐색하여, U로부터 가장 가까운 프로토타입을 가지는 x를 찾는다.
- X로부터 x를 삭제하고 그것을 U에 더한다.
- U에 더 이상 프로토타입을 더할 수 없을 때까지 탐색을 반복한다.

그렇다면 어떻게 y의 라벨을 찾을 수 있을까? 다음과 같은 수식을 따른다

a(x) = || x' - y || / || x - y ||

- 위의 수식은 "경계율"이라 부름
- ||x - y|| 는 x와 다른 라벨을 갖는 데이터 최근접 데이터 y와의 거리
- ||x' - y|| 는 x와 같은 라벨을 갖는 데이터 중 y에 가장 가까운 x'까지의 거리
- 경계율의 범위는 0과 1 사이임
- 1보다 커질 수 없는 이유는 ||x - y||가 ||x' - y|| 보다 절대 클 수 없기 때문임
- 확률값이라고 생각하면 편함!
